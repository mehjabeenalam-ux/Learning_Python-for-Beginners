ğŸš Mini Project 2: Drone Swarm Simulation (Multiprocessing)



ğŸ¯ Goal: Simulate multiple drones flying & spraying simultaneously

```python
from multiprocessing import Process
import time

def drone_mission(drone_id, area_acres, battery_percent):
    print(f"ğŸš Drone {drone_id} taking off")
    time.sleep(1)

    spray_time = area_acres * 2
    battery_used = area_acres * 5
    battery_left = battery_percent - battery_used

    print(f"ğŸŒ¾ Drone {drone_id} sprayed {area_acres} acres")
    print(f"ğŸ”‹ Battery left: {battery_left}%")

if __name__ == "__main__":
    drones = [
        ("D1", 2, 100),
        ("D2", 3, 100),
        ("D3", 1.5, 100),
        ("D4", 2.5, 100)
    ]

    processes = []

    for drone in drones:
        p = Process(target=drone_mission, args=drone)
        processes.append(p)
        p.start()

    for p in processes:
        p.join()

    print("âœ… All drone missions completed")
```
