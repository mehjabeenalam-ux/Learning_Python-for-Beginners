Basic Multiprocessing Syntax

```python
from multiprocessing import Process

def task():
    print("Task running in parallel")

p1 = Process(target=task)
p2 = Process(target=task)

p1.start()
p2.start()

p1.join()
p2.join()
```

Cooking 2 italian Dishes Simulatenously Example

```python
from multiprocessing import Process

def task():
    print("Chefs cooking Italian dishes pasta and lasagna simaultaneously!")

pasta_cooking = Process(target=task)
lasangna_cooking = Process(target=task)

pasta_cooking.start()
lasangna_cooking.start()

pasta_cooking.join()
lasangna_cooking.join()
```

Chopping Onions and Frying Spices

```python
from multiprocessing import Process

def task():
    print("Chef is chopping onions and frying spices at the same time!")

chopping_onions = Process(target=task)
frying_spices = Process(target=task)

chopping_onions.start()
frying_spices.start()

chopping_onions.join()
frying_spices.join()
```

✔ Each process runs independently on a different CPU core.

Multiprocessing Pool Squaring numbers in a given sequence

```python
from multiprocessing import Pool

def square(n):
    return n * n

with Pool(4) as p:   # 4 CPU cores
    result = p.map(square, [1, 2, 3, 4])

print(result)
```

✔ Distributes tasks automatically across multiple processes.
