Basic Shallow Copy Exercise

```python
original_dict = {'a': 1, 'b': [2, 3]}
shallow_dict = original_dict.copy()  # Using dict's copy method for shallow copy

# Modify the nested list in shallow_dict
shallow_dict['b'][0] = 'oops'

# Print both and observe
print(original_dict)
print(shallow_dict)
```

Deep Copy with Custom Objects Exercise

```python
import copy

class Person:
    def __init__(self, name, friends):
        self.name = name
        self.friends = friends

alice = Person('Alice', ['Bob'])
group = [alice]
deep_group = copy.deepcopy(group)

# Add a friend to the deep copy's Alice
deep_group[0].friends.append('Charlie')

# Print originals vs deep
print(group[0].friends)      # Should be ['Bob']
print(deep_group[0].friends) # Should be ['Bob', 'Charlie']
```

- Run this to compare execution times; shallow is faster for big data when nested independence isn't needed, highlighting performance trade-offs in projects.

```python
import copy
import time

large_list = [[i] * 1000 for i in range(1000)]  # Nested large data

start = time.time()
shallow = copy.copy(large_list)
print("Shallow time:", time.time() - start)

start = time.time()
deep = copy.deepcopy(large_list)
print("Deep time:", time.time() - start)
```
