```python
import time

def bake_pizza(pizza_name):
    print(f"üçï Preparing {pizza_name}")
    time.sleep(2)  # preparing
    print(f"üî• Baking {pizza_name}")
    time.sleep(3)  # baking
    print(f"‚úÖ {pizza_name} ready\n")

start = time.time()

bake_pizza("Margherita")
bake_pizza("Pepperoni")
bake_pizza("Veggie")

print("‚è±Ô∏è Total time:", round(time.time() - start, 2), "seconds")
```
