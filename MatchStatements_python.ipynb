Match Statements in Python

A match statement (introduced in Python 3.10) is a structural pattern-matching statement that allows you to compare a value against multiple patterns, similar to a switch-case in other languages, but more powerful.



https://www.w3schools.com/python/python_match.asp

Instead of writing many if..else statements, you can use the match statement.

The match statement selects one of many code blocks to be executed.

Basic Match Statement Exercises

Exercise 1: Matching Day

```python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
```

Default Value

Use the underscore character _ as the last case value if you want a code block to execute when there are not other matches:

```python
day = 4
match day:
  case 6:
    print("Today is Saturday")
  case 7:
    print("Today is Sunday")
  case _:
    print("Looking forward to the Weekend")
```

Combine Values

Use the pipe character | as an or operator in the case evaluation to check for more than one value match in one case:

```python
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
```

If Statements as Guards

You can add if statements in the case evaluation as an extra condition-check:

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

Match Statement for a Web App with Multiple Functions

```python
action = {"type": "purchase", "user": "Alice", "item": "Laptop"}

match action:
    case {"type": "login", "user": user}:
        print(f"{user} logged in")
    case {"type": "logout", "user": user}:
        print(f"{user} logged out")
    case {"type": "purchase", "user": user, "item": item}:
        print(f"{user} purchased {item}")
    case {"type": "comment", "user": user, "comment": cmt}:
        print(f"{user} commented: {cmt}")
    case _:
        print("Unknown action")

```

Match Statement with User Inputs

```python
# Match Statements Example
action_type = input("Enter action type (login/logout/purchase/comment): ").strip().lower()
user = input("Enter username: ").strip()

if action_type == "purchase":
    item = input("Enter item name: ").strip()
    action = {"type": action_type, "user": user, "item": item}
elif action_type == "comment":
    comment = input("Enter comment: ").strip()
    action = {"type": action_type, "user": user, "comment": comment}
else:
    action = {"type": action_type, "user": user}

# Match statement to handle actions
match action:
    case {"type": "login", "user": user}:
        print(f"{user} logged in")
    case {"type": "logout", "user": user}:
        print(f"{user} logged out")
    case {"type": "purchase", "user": user, "item": item}:
        print(f"{user} purchased {item}")
    case {"type": "comment", "user": user, "comment": cmt}:
        print(f"{user} commented: {cmt}")
    case _:
        print("Unknown action")
```
