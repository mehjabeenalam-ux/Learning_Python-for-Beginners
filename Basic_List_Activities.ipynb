Accessing Lists

```python
employee_list = ["Amy", "John", "Mary", "Ken", "Mary"]

#Accessing Items
print(employee_list[0])
print(employee_list[1])
print(employee_list[2])
print(employee_list[3])
print(employee_list[4])

print(employee_list[-1])
print(employee_list[-2])
print(employee_list[-3])
print(employee_list[-4])
print(employee_list[-5])

#Slicing

print(employee_list[0:2]) #First 2 elements
print(employee_list[0:]) #Entire List
print(employee_list[1:])
print(employee_list[0:7])

#Len
print(len(employee_list))
```


Changing items in a list

```python
employee_list = ["Amy", "John", "Mary", "Ken", "Mary"]

employee_list[0] = "Mehjabeen"
employee_list[4] = "Nazmul"
employee_list[-1] = "Nazmul"
print(employee_list)
```

Add items to a list to the end

```python
employee_list.append("Fatema")

print(employee_list)
```

Inserting Items in a List in a specific position

```python
employee_list.insert(0, "Tareq")
print(employee_list)
```

Remove Items from a List

```python
# Remove by value

employee_list.remove("Nazmul")
print(employee_list)

# Remove and return last

employee_list.pop()
print(employee_list)

del employee_list[0]
print(employee_list)
```


List methods

```python
# Information
numbers = [3, 1, 4, 1, 5, 9]

print(len(numbers))         # 6 (length)
print(numbers.count(1))     # 2 (count occurrences)
print(numbers.index(4))     # 2 (find position)
print(numbers.index(3))
print(numbers.index(1))
print(numbers.index(9))
```
