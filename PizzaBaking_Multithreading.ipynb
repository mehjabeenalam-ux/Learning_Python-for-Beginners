üçï 2Ô∏è‚É£ Mini Project: Multithreading (Multiple Chefs, One Kitchen)
üß† Concept

Multiple chefs (threads)

Same kitchen (same process)

Great for waiting tasks like baking

```python
import threading
import time

def bake_pizza(pizza_name):
    print(f"üçï Preparing {pizza_name}")
    time.sleep(2)
    print(f"üî• Baking {pizza_name}")
    time.sleep(3)
    print(f"‚úÖ {pizza_name} ready\n")

start = time.time()

threads = []
pizzas = ["Margherita", "Pepperoni", "Veggie"]

for pizza in pizzas:
    t = threading.Thread(target=bake_pizza, args=(pizza,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("‚è±Ô∏è Total time:", round(time.time() - start, 2), "seconds")
```
