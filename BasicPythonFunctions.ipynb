Basic Python Functions

Length of a variable

```python
print(len('Python'))
```

Convert a number to an Integer

```python
print(int(3.99))
```

Convert a number to a float

```python
print(float(3))
```

Createing Ranges

```python
print(range(5))

for x in range(5):
  print(x)

print(range(1,11,1))

for i in range(1,11,1):
  print(i)
```

Variable Type Checking

```python
print(type(3))
print(type(3.14))
print(type("3"))
print(type("True"))
```
