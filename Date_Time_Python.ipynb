2. Create Date and Time Objects: Custom Date Time


You can create specific date or time objects by passing integer arguments to the constructors.

```python
from datetime import date, time, datetime

# Create a specific date: Year, month, day are mandatory
my_date = date(2020, 5, 17)
print(f"My date: {my_date}")

# Create a specific time: Hour, minute, second are optional
my_time = time(12, 30, 45)
print(f"My time: {my_time}")

# Create a specific datetime
my_datetime = datetime(2020, 5, 17, 12, 30, 45)
print(f"My datetime: {my_datetime}")
```

3. Format Dates and Times into Strings


The strftime() method is used to create human-readable strings from datetime objects using format codes.

Creating Date Objects

- To create a date, we can use the datetime() class (constructor) of the datetime module.

- The datetime() class requires three parameters to create a date: year, month, day.

```python
from datetime import datetime

now = datetime.now()
# Format the date as "Tuesday, January 06, 2025, 04:19 PM"
formatted_date = now.strftime("%A, %B %d, %Y, %I:%M %p")
print(formatted_date)
```

4. Perform Date Arithmetic

Use the timedelta class to calculate the difference between dates or add/subtract time intervals.

```python
from datetime import datetime, timedelta

# Two dates
date1 = datetime(2025, 1, 1)
date2 = datetime(2025, 1, 31)

# Calculate the difference (returns a timedelta object)
difference = date2 - date1
print(f"Difference in days: {difference.days}") # Output: 30

# Add days to a date
future_date = date1 + timedelta(days=10, hours=2)
print(f"Date after 10 days and 2 hours: {future_date}")
# Output: 2025-01-11 02:00:00
```
