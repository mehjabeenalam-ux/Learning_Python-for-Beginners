Split Function

```python
# Mini Project: Simple score parser
scores_text = """
Alice,95,Math
Bob,82,Science
Charlie,78,History
"""

lines = scores_text.strip().split("\n")

for line in lines:
    name, score, subject = line.split(",")
    print(f"{name} got {score} in {subject}")
    # Output:
    # Alice got 95 in Math
    # Bob got 82 in Science
    # Charlie got 78 in History
```

Slicing Function

```python
# Mini Project: Show last 5 log entries
log_lines = [
    "2025-01-01 INFO Start",
    "2025-01-02 ERROR Database timeout",
    "2025-01-03 INFO User login",
    "2025-01-04 WARN High memory",
    "2025-01-05 ERROR Crash",
    "2025-01-06 INFO Success"
]

last_five = log_lines[-5:]          # very clean & readable

for line in last_five:
    print(line)
```

Mini Project Combining Split and Slice Functions

```python
# Mini Project: Process CSV line and get first 3 columns
line = "2025-01-05,Tokyo,25.5,Cloudy,80"

# split + slicing together
fields = line.split(",")
date, city, temp = fields[:3]          # clean slicing after split

print(f"On {date} in {city} was {temp}°C")
```
