1. Basic Level
Write a function that returns different messages based on temperature:
< 0 → "Freezing"
0–15 → "Cold"
16–25 → "Pleasant"25 → "Hot"

```python
def describe_temperature(temp):
    match temp:
        case t if t < 0:
            return "Freezing"
        case t if 0 <= t <= 15:
            return "Cold"
        case t if 16 <= t <= 25:
            return "Pleasant"
        case t if t > 25:
            return "Hot"
        case _:
            return "Invalid temperature"

# Test
print(describe_temperature(-5))   # Freezing
print(describe_temperature(10))   # Cold
print(describe_temperature(20))   # Pleasant
print(describe_temperature(32))   # Hot
```

```python
#Alterantive Style

def describe_temperature(temp):
    match temp:
        case _ if temp < 0:
            return "Freezing"
        case _ if temp <= 15:
            return "Cold"
        case _ if temp <= 25:
            return "Pleasant"
        case _:
            return "Hot"

print(describe_temperature(-5))
print(describe_temperature(15))
print(describe_temperature(24))
```

2. Medium Level
Create a function describe_shape(shape) that takes a tuple:
(length,) → "Line"
(length, width) → "Rectangle"
(radius,) → "Circle"
(a, b, c) → "Triangle"
anything else → "Unknown shape"

```python
def describe_shape(shape):
    match shape:
        case (length,):
            return "Line"
        case (length, width):
            return "Rectangle"
        case (radius,):
            return "Circle"
        case (a, b, c):
            return "Triangle"
        case _:
            return "Unknown shape"

# Test cases
print(describe_shape((5,)))          # Line
print(describe_shape((4, 6)))        # Rectangle
print(describe_shape((3.5,)))        # Circle
print(describe_shape((3, 4, 5)))     # Triangle
print(describe_shape((1, 2, 3, 4)))  # Unknown shape
```

3. Harder Level
Write a mini calculator function that takes a list:
["add", 5, 3] → 8
["subtract", 10, 4] → 6
["power", 2, 3] → 8
Any invalid format or unknown operation → "Invalid"

```python
def mini_calculator(command):
    match command:
        case ["add", x, y]:
            return float(x) + float(y)
        case ["subtract", x, y]:
            return float(x) - float(y)
        case ["multiply", x, y]:
            return float(x) * float(y)
        case ["divide", x, y] if float(y) != 0:
            return float(x) / float(y)
        case ["divide", x, y]:
            return "Cannot divide by zero"
        case ["power", x, y]:
            return float(x) ** float(y)
        case _:
            return "Invalid command or format"

# Test cases
print(mini_calculator(["add", 5, 3]))          # 8.0
print(mini_calculator(["subtract", 10, 4]))    # 6.0
print(mini_calculator(["multiply", 2, 3]))     # 6.0
print(mini_calculator(["power", 2, 3]))        # 8.0
print(mini_calculator(["divide", 10, 0]))      # Cannot divide by zero
print(mini_calculator(["add", 5]))             # Invalid command or format
print(mini_calculator("hello"))                # Invalid command or format
```

Alternative Code to the one above

```python
def mini_calculator_safe(command):
    match command:
        case [op, x, y] if op in ("add", "subtract", "multiply", "divide", "power"):
            try:
                x, y = float(x), float(y)
                match op:
                    case "add":
                        return x + y
                    case "subtract":
                        return x - y
                    case "multiply":
                        return x * y
                    case "divide" if y != 0:
                        return x / y
                    case "divide":
                        return "Division by zero"
                    case "power":
                        return x ** y
            except ValueError:
                return "Invalid numbers"
        case _:
            return "Invalid command format"

# Test with valid list arguments
print(mini_calculator_safe(["add", 1, 2]))
print(mini_calculator_safe(["subtract", 5, 3]))
print(mini_calculator_safe(["multiply", 4, 2]))
print(mini_calculator_safe(["divide", 10, 2]))
print(mini_calculator_safe(["power", 2, 3]))
print(mini_calculator_safe(["divide", 5, 0]))
print(mini_calculator_safe(["add", "a", 2]))
print(mini_calculator_safe(["unknown", 1, 2]))
print(mini_calculator_safe(1)) # This would now return "Invalid command format"
print(mini_calculator_safe(["add", 1])) # This would also return "Invalid command format"
```
