We are showing 2 Scenarios for a factory to meet its energy requirements from solar, wind and grid energy.

Scenario 1 shows daily factory energy requirement of 50 kwH against which 59.5 kwH is generated from these three sources resulting in surplus energy.

Scenario 2shows daily factory energy requirement of 70 kwH against which 59.5 kwH is generated from these three sources resulting in deficit energy availability.

Scenario 1 : Suplus Energy Production

```python
# Parent class
class EnergySource:
    def __init__(self, name):
        self.name = name

    def generate_energy(self):
        print("Generating energy...")


# Solar energy class (Dhaka-based calculation)
class SolarEnergy(EnergySource):
    def generate_energy(self):
        panel_capacity_kw = 5      # 5 kW rooftop solar system
        dhaka_sun_hours = 5.5      # Average peak sun hours in Dhaka
        energy = panel_capacity_kw * dhaka_sun_hours
        print(f"‚òÄÔ∏è Solar generated {energy} kWh today")


# Wind energy class
class WindEnergy(EnergySource):
    def generate_energy(self):
        energy = 12  # kWh (average small wind turbine output)
        print(f"üå¨Ô∏è Wind generated {energy} kWh today")


# Grid energy class
class GridEnergy(EnergySource):
    def generate_energy(self):
        energy = 20  # kWh supplied from grid
        print(f"‚ö° Grid supplied {energy} kWh today")


# Function demonstrating polymorphism
def daily_energy_report(source):
    source.generate_energy()


# Create objects
solar = SolarEnergy("Solar")
wind = WindEnergy("Wind")
grid = GridEnergy("Grid")

# Store all energy sources
energy_sources = [solar, wind, grid]

# Generate daily report
for source in energy_sources:
    daily_energy_report(source)

# Total energy calculation
total_energy = (5 * 5.5) + 12 + 20
print(f"\nüîã Total energy available today: {total_energy} kWh")

# Factory electricity requirement
factory_daily_requirement = 50  # kWh needed per day

print(f"\nüè≠ Factory daily electricity requirement: {factory_daily_requirement} kWh")

# Compare generation vs requirement
if total_energy >= factory_daily_requirement:
    surplus = total_energy - factory_daily_requirement
    print(f"‚úÖ Energy requirement met!")
    print(f"üîã Surplus energy available: {surplus} kWh")
else:
    deficit = factory_daily_requirement - total_energy
    print(f"‚ö†Ô∏è Energy shortfall!")
    print(f"‚ö° Additional {deficit} kWh needed from grid or storage")

```

Scenario 2: Deficit Energy Generation

```python
# Factory electricity requirement
factory_daily_requirement = 70  # kWh needed per day

print(f"\nüè≠ Factory daily electricity requirement: {factory_daily_requirement} kWh")

# Compare generation vs requirement
if total_energy >= factory_daily_requirement:
    surplus = total_energy - factory_daily_requirement
    print(f"‚úÖ Energy requirement met!")
    print(f"üîã Surplus energy available: {surplus} kWh")
else:
    deficit = factory_daily_requirement - total_energy
    print(f"‚ö†Ô∏è Energy shortfall!")
    print(f"‚ö° Additional {deficit} kWh needed from grid or storage")
```
