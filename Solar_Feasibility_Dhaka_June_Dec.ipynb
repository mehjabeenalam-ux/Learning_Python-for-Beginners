Feasibility of Solar Panels for a household and a Factory in Dhaka during June and December months in Bangladesh

```python
from abc import ABC, abstractmethod

# Abstract Energy Source
class EnergySource(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def generate_energy(self):
        pass

# Solar Energy Class
class SolarEnergy(EnergySource):
    def __init__(self, name, capacity_kw, sun_hours_per_day):
        super().__init__(name)
        self.capacity_kw = capacity_kw
        self.sun_hours_per_day = sun_hours_per_day

    def generate_energy(self):
        """Daily energy in kWh"""
        energy = self.capacity_kw * self.sun_hours_per_day
        print(f"{self.name} generates {energy:.2f} kWh/day")
        return energy

# Household and Factory Simulation
def monthly_simulation(entity_name, daily_requirement, solar: SolarEnergy, month_days=30, cost_per_kwh=8):
    daily_solar = solar.generate_energy()
    total_solar = daily_solar * month_days
    total_requirement = daily_requirement * month_days
    supplied_by_solar = min(total_solar, total_requirement)
    grid_needed = max(0, total_requirement - total_solar)
    savings = supplied_by_solar * cost_per_kwh

    print(f"\n--- {entity_name} ---")
    print(f"Monthly energy requirement: {total_requirement} kWh")
    print(f"Total solar contribution: {supplied_by_solar:.2f} kWh")
    print(f"Energy from grid: {grid_needed:.2f} kWh")
    print(f"Electricity bill savings: {savings:.2f} Tk")

# June Simulation
print("ðŸŒž Simulation for June (Summer)")
house_solar_june = SolarEnergy("Household Solar", 3, 6)
factory_solar_june = SolarEnergy("Factory Solar", 10, 6)

monthly_simulation("Household", 10, house_solar_june)
monthly_simulation("Factory", 50, factory_solar_june)

# December Simulation
print("\nâ„ï¸ Simulation for December (Winter)")
house_solar_dec = SolarEnergy("Household Solar", 3, 4.5)
factory_solar_dec = SolarEnergy("Factory Solar", 10, 4.5)

monthly_simulation("Household", 10, house_solar_dec)
monthly_simulation("Factory", 50, factory_solar_dec)
```
