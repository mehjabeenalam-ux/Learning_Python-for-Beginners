Drone Swarm Spray Project Without OOP and Inheritance

```python
# -------- Drone 1: Maize --------
drone1_id = "M1"
drone1_battery = 100
drone1_payload = 15
drone1_area = 3

print(f"Drone {drone1_id} is flying")
fertilizer1 = drone1_area * 1.2
print(f"Spraying {fertilizer1} liters on maize")
print(f"Drone {drone1_id} returning to base")

print("--------------------------------")

# -------- Drone 2: Maize --------
drone2_id = "M2"
drone2_battery = 100
drone2_payload = 15
drone2_area = 3

print(f"Drone {drone2_id} is flying")
fertilizer2 = drone2_area * 1.2
print(f"Spraying {fertilizer2} liters on maize")
print(f"Drone {drone2_id} returning to base")

print("--------------------------------")

# -------- Drone 3: Tomato --------
drone3_id = "T1"
drone3_battery = 100
drone3_payload = 10
drone3_area = 4

print(f"Drone {drone3_id} is flying")
fertilizer3 = drone3_area * 0.8
print(f"Spraying {fertilizer3} liters on tomato")
print(f"Drone {drone3_id} returning to base")
```
