Python does not have built-in support for Arrays, but Python Lists can be used instead.

An array is a special variable, which can hold more than one value at a time.

If you have a list of items (a list of car names, for example), storing the cars in single variables will look loike this:

car1 = "Ford"
car2 = "Volvo"
car3 = "BMW"

However, what if you want to loop through the cars and find a specific one? And what if you had not 3 cars, but 300?

The solution is an array!

An array can hold many values under a single name, and you can access the values by referring to an index number.

Exercises using Python Lists

üîπ Exercise 1: Student Scores (Dynamic Data)

```python
scores = [85, 90, 78, 92]

# Add a new score
scores.append(88)

# Remove a score
scores.remove(78)

# Average score
average = sum(scores) / len(scores)
print("Average:", average)
```

‚úÖ Why List?

Dynamic size

Easy insertion/removal

```python
cart = ["Laptop", "Mouse", "Keyboard"]

if "Mouse" in cart:
    print("Item already added")

cart.append("Headphones")
print(cart)
```

üîπ Exercise 3: Sensor Temperature Readings

```python
from array import array

temps = array('i', [30, 32, 31, 29, 28])

print("First reading:", temps[0])
print("Total readings:", len(temps))
```

‚úÖ Why Array?

Memory-efficient

Same data type

üîπ Exercise 4: Fixed Exam Marks

```python
from array import array

marks = array('i', [70, 75, 80, 85])

marks[2] = 90
print(marks)
```

üîπ Exercise 5: Fast Math Operations

```python
import numpy as np

data = np.array([10, 20, 30, 40])

# Vectorized operation
result = data * 2
print(result)
```

üöÄ Why NumPy?

No loops needed

Extremely fast

üîπ Exercise 6: Monthly Sales Analysis

```python
sales = np.array([2000, 2500, 3000, 2800])

print(sales)

print("Total Sales:", np.sum(sales))
print("Average Sales:", np.mean(sales))
```

üåç Real use: Finance, business analytics

üîπ Exercise 7: Image Pixel Matrix (2D Array)

```python
image = np.array([
    [255, 200, 180],
    [120, 100, 90],
    [60,  30,  10]
])

print("Pixel at (1,1):", image[1,1])
```

üåç Real use: Image processing, computer vision
