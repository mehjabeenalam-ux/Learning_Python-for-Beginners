Execution Time and Memory Usage of Lists, Sets, Tuples, Dictionaries, Arrays, NumPy Arrays and Iterators in Python

```python
import time
import sys
from array import array
import numpy as np

# -----------------------------
# SAME DATASET (Large)
# -----------------------------
N = 100_000
data = list(range(N))

structures = {
    "List": data,
    "Set": set(data),
    "Tuple": tuple(data),
    "Dictionary": {i: i for i in data},
    "Array": array('i', data),
    "NumPy Array": np.array(data),
    "Iterator": iter(data)
}

print(f"{'Structure':<12} | {'Time (s)':<10} | {'Memory (bytes)'}")
print("-" * 45)

# -----------------------------
# BENCHMARK
# -----------------------------
for name, obj in structures.items():

    start = time.perf_counter()
    total = 0

    if name == "NumPy Array":
        total = obj.sum()            # vectorized
    else:
        for x in obj:
            total += x               # iteration

    end = time.perf_counter()

    memory = sys.getsizeof(obj)

    print(f"{name:<12} | {end-start:<10.6f} | {memory}")
```
