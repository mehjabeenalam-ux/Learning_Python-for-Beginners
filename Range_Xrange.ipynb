Range Vs Ex Range in Python

range() ‚Üí returns a list (stores all values in memory)

xrange() ‚Üí returns a generator-like object (creates values on demand)

üìå In Python 3:

range() is optimized and now behaves like old xrange()

xrange() ‚ùå does not exist

Both are memory efficient.

Range is an object type and exists in Python 3 versions and works best on loops and iterations.

xRange existed in legacy Python 2 version systems and is an x range object type.

Memory Comparison between Range and List

```python
import sys

r = range(1000000)
l = list(range(1000000))

print("range memory:", sys.getsizeof(r))
print("list memory:", sys.getsizeof(l))
```

Range occupies way less memory compared to a list as it generates a sequence only when needed!
