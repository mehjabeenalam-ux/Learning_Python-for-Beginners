```python
#ðŸ§± Step 1: Parent Class (Base Class)



class Animal:

  def eat(self):

    print("The animal is eating")



#Here in Step 1 we have created a class or a blueprint for Animal with a method eat.



#ðŸ§± Step 2: Child Class (Inheritance)



class Dog(Animal):

  def bark(self):

    print("The dog is barking")



#Here the child class Dog inherits from the parent class Animal.



#ðŸ§± Step 3: Create an Object



my_dog = Dog()



#Here my_dog is an object created from the Dog class.



#ðŸ§± Step 4: Use Inherited Method



my_dog.eat()  # inherited from Animal

my_dog.bark() # defined in Dog

```
