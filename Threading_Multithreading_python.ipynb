Threading Vs Multithreading in Python

Threading: Basic threading means running tasks one after another using one thread (even if you use the threading module).

Single Threading Basic Syntax

```python
def task():
    print("Task running")

task()
task()
```

Solar Sensor Single Threading

```python
import time

def read_sensor(sensor_id):
    print(f"Reading sensor {sensor_id}")
    time.sleep(2)

read_sensor(1)
read_sensor(2)
read_sensor(3)
```

Multithreading: Multithreading means running multiple threads concurrently so tasks appear to happen at the same time.

Basic Syntax Multithreading

```python
import threading

def task():
    print("Task running")

t1 = threading.Thread(target=task)
t2 = threading.Thread(target=task)

t1.start()
t2.start()
```

Solar Sensors on parallel I/O

```python
import threading
import time

def read_sensor(sensor_id):
    print(f"Reading sensor {sensor_id}")
    time.sleep(2)

threads = []

for i in range(1, 4):
    t = threading.Thread(target=read_sensor, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```
