```python
from functools import reduce
```

ðŸ”¹ Step 1: Create Weekly Sales Data

```python
weekly_sales = [50, 120, 80, 200, 150, 90, 300]

print(weekly_sales)
```

ðŸ§  Explanation

Each number represents sales for one day of the week.

ðŸ”¹ Step 2: FILTER â€“ Remove Low Sales

ðŸ‘‰ Rule: Keep sales â‰¥ 100

```python
filtered_sales = list(filter(lambda sale: sale >= 100, weekly_sales))
print("Filtered Sales:", filtered_sales)
```

ðŸ§  Explanation (Very Simple)

filter() â†’ keeps only values that match a rule

lambda sale: sale >= 100 â†’ rule for filtering

ðŸ”¹ Step 3: MAP â€“ Apply Bonus (Double Sales)

```python
bonus_sales = list(map(lambda sale: sale * 2, filtered_sales))
print("Bonus Sales:", bonus_sales)
```

ðŸ§  Explanation

map() â†’ changes every value in the list

We double sales to simulate bonus or promotion

ðŸ”¹ Step 4: REDUCE â€“ Calculate Total Revenue

```python
total_revenue = reduce(lambda a, b: a + b, bonus_sales)
print("Total Revenue:", total_revenue)
```

ðŸ§  Explanation

reduce() â†’ combines all numbers into one value

Adds all bonus sales together

```python
from functools import reduce

# Step 1: Weekly sales data
weekly_sales = [50, 120, 80, 200, 150, 90, 300]

# Step 2: Filter low sales
filtered_sales = list(filter(lambda sale: sale >= 100, weekly_sales))
print("Filtered Sales:", filtered_sales)

# Step 3: Apply bonus using map
bonus_sales = list(map(lambda sale: sale * 2, filtered_sales))
print("Bonus Sales:", bonus_sales)

# Step 4: Calculate total revenue using reduce
total_revenue = reduce(lambda a, b: a + b, bonus_sales)
print("Total Revenue:", total_revenue)

# Step 5: Highest sale day
highest_sale = max(bonus_sales)
print("Highest Sale Day:", highest_sale)
```

```python
from functools import reduce

def is_high_sale(sale):
    return sale >= 100

def apply_bonus(sale):
    return sale * 2

def add_sales(a, b):
    return a + b

weekly_sales = [50, 120, 80, 200, 150, 90, 300]

filtered_sales = list(filter(is_high_sale, weekly_sales))
bonus_sales = list(map(apply_bonus, filtered_sales))
total_revenue = reduce(add_sales, bonus_sales)
highest_sale = max(bonus_sales)

print("Filtered Sales:", filtered_sales)
print("Bonus Sales:", bonus_sales)
print("Total Revenue:", total_revenue)
print("Highest Sale Day:", highest_sale)
```
