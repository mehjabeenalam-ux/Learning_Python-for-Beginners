Swap variables allow instant, safe, and readable switching between energy sourcesâ€”ideal for smart energy systems.

Scenario 1: Complete Reliance on Solar Energy

```python
solar = "OFF"
grid = "ON"

solar_energy_generated = 18      # kWh/day
total_energy_required = 8.8      # kWh/day

time_of_day = "DAY"   # DAY or NIGHT

match time_of_day:
    case "DAY":
        solar, grid = grid, solar
        energy_from_solar = min(solar_energy_generated, total_energy_required)
        energy_from_grid = total_energy_required - energy_from_solar

    case "NIGHT":
        energy_from_solar = 0
        energy_from_grid = total_energy_required

print("ðŸ”Œ Power Status")
print("Solar:", solar)
print("Grid:", grid)

print("\nâš¡ Energy Distribution (kWh)")
print("Solar Contribution:", energy_from_solar)
print("Grid Contribution:", energy_from_grid)
```
