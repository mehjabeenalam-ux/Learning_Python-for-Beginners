üöÅ Drone Swarm Fertilizer Spraying Project (Python OOP + Inheritance + Polymorphism)

üß© Step 1: Base Drone Class (Parent)

```python
class Drone:
    def __init__(self, drone_id, battery, payload_capacity):
        self.drone_id = drone_id
        self.battery = battery          # %
        self.payload_capacity = payload_capacity  # liters

    def fly(self):
        print(f"Drone {self.drone_id} is flying")

    def spray(self, area):
        print("Spraying fertilizer")

    def return_home(self):
        print(f"Drone {self.drone_id} returning to base")
```

üîç What this represents

This is a generic drone:

Can fly

Can spray

Can return


üß© Step 2: Inheritance ‚Äì Crop-Specific Drones

```python
class MaizeDrone(Drone):
    def spray(self, area):
        fertilizer = area * 1.2  # liters per acre
        print(f"MaizeDrone spraying {fertilizer} liters over {area} acres")


class TomatoDrone(Drone):
    def spray(self, area):
        fertilizer = area * 0.8  # liters per acre
        print(f"TomatoDrone spraying {fertilizer} liters over {area} acres")

```

üî• POLYMORPHISM (MOST IMPORTANT PART)

üëâ Same method name: spray()
üëâ Different behavior depending on drone type


```python
def start_mission(drone, area):
    drone.fly()
    drone.spray(area)
    drone.return_home()
```

üß† This works because:

Python decides at runtime which spray() to run.

üß© Step 3: Drone Swarm Mission
üåæ Land Distribution

Total land = 10 acres

üåΩ Maize = 6 acres

üçÖ Tomato = 4 acres

```python
maize_drone_1 = MaizeDrone("M1", 100, 15)
maize_drone_2 = MaizeDrone("M2", 100, 15)
tomato_drone = TomatoDrone("T1", 100, 10)

start_mission(maize_drone_1, 3)
start_mission(maize_drone_2, 3)
start_mission(tomato_drone, 4)
```
