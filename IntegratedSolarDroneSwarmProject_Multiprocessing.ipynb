üåûüöÅ Integrated Solar + Drone Energy System

Solar Plants (Energy Producers)
        
        ‚Üì

   Shared Energy Pool

        ‚Üì
        
Drone Swarm (Energy Consumers)


```python
from multiprocessing import Process, Manager
import time

def solar_plant(plant_id, capacity_kw, sun_hours, energy_pool):
    print(f"üåû Solar Plant {plant_id} generating energy...")
    time.sleep(1)

    energy_generated = capacity_kw * sun_hours
    energy_pool.append(energy_generated)

    print(f"‚úÖ Plant {plant_id} produced {energy_generated} kWh")

def drone_mission(drone_id, area_acres, energy_pool):
    print(f"üöÅ Drone {drone_id} starting mission")
    time.sleep(1)

    energy_needed = area_acres * 3  # kWh per acre

    if sum(energy_pool) >= energy_needed:
        energy_pool[0] -= energy_needed
        print(f"üåæ Drone {drone_id} sprayed {area_acres} acres")
        print(f"üîã Energy used: {energy_needed} kWh")
    else:
        print(f"‚ùå Drone {drone_id}: Not enough solar energy")

if __name__ == "__main__":
    manager = Manager()
    energy_pool = manager.list()

    # üåû Solar plants
    solar_plants = [
        ("S1", 5, 6),
        ("S2", 8, 5.5),
        ("S3", 4, 6)
    ]

    solar_processes = []
    for plant in solar_plants:
        p = Process(target=solar_plant, args=(*plant, energy_pool))
        solar_processes.append(p)
        p.start()

    for p in solar_processes:
        p.join()

    print(f"\n‚ö° Total Solar Energy Available: {sum(energy_pool)} kWh\n")

    # üöÅ Drone swarm
    drones = [
        ("D1", 2),
        ("D2", 1.5),
        ("D3", 2.5),
        ("D4", 1)
    ]

    drone_processes = []
    for drone in drones:
        p = Process(target=drone_mission, args=(*drone, energy_pool))
        drone_processes.append(p)
        p.start()

    for p in drone_processes:
        p.join()

    print(f"\nüîã Remaining Energy: {sum(energy_pool)} kWh")
```
