üß† Concept

Multiple kitchens (processes)

Each has its own oven & chef

True parallel cooking

```python
import multiprocessing
import time

def bake_pizza(pizza_name):
    print(f"üçï {pizza_name} started")
    time.sleep(5)
    print(f"‚úÖ {pizza_name} ready\n")

start = time.time()

processes = []
pizzas = ["Margherita", "Pepperoni", "Veggie"]

for pizza in pizzas:
    p = multiprocessing.Process(target=bake_pizza, args=(pizza,))
    processes.append(p)
    p.start()

for p in processes:
    p.join()

print("‚è±Ô∏è Total time:", round(time.time() - start, 2), "seconds")
```
