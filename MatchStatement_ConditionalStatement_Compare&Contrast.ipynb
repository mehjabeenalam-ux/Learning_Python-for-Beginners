2 Exercises to Compare and Contrast the Superiority of Match Statements and Conditional Statements

Exercise 1: Problem: Determine a personâ€™s eligibility based on age

Dataset: A single user input for age


Why Conditional Statements Are Superior Here:

- Only numeric comparisons

- Very few conditions

- Easy, readable, works in all Python versions

- Using a match statement would be overkill

Coding Using Conditional Statements

```python
# Input
age = int(input("Enter your age: "))

# Conditional statement to check eligibility
if age < 0:
    print("Invalid age")
elif age < 18:
    print("Minor: Cannot vote or drink")
elif 18 <= age < 21:
    print("Can vote, cannot drink")
else:
    print("Adult: Can vote and drink")
```

Coding Using Match Statement

```python
# Using match (Python 3.10+)
age = int(input("Enter your age: "))

match age:
    case age if age < 0:
        print("Invalid age")
    case age if age < 18:
        print("Minor: Cannot vote or drink")
    case age if 18 <= age < 21:
        print("Can vote, cannot drink")
    case _:
        print("Adult: Can vote and drink")
```

Verdict: Stick to Conditional Statement in this case instead of Match Statement

Exercise 2: Problem: Handling user actions in a web app

Dataset: A dictionary representing a user action: login, logout, purchase, or comment

Goal: Print a message depending on the action type

Coding Using Conditional Statement

```python
# Input
action_type = input("Enter action type (login/logout/purchase/comment): ").strip().lower()
user = input("Enter username: ").strip()

if action_type == "login":
    print(f"{user} logged in")
elif action_type == "logout":
    print(f"{user} logged out")
elif action_type == "purchase":
    item = input("Enter item name: ").strip()
    print(f"{user} purchased {item}")
elif action_type == "comment":
    comment = input("Enter comment: ").strip()
    print(f"{user} commented: {comment}")
else:
    print("Unknown action")
```

Drawbacks:

Multiple nested if-elif blocks

Hard to extend if more action types are added

Less readable with structured data

Coding Using Match Statement

```python
# Input
action_type = input("Enter action type (login/logout/purchase/comment): ").strip().lower()
user = input("Enter username: ").strip()

# Construct action dictionary
if action_type == "purchase":
    item = input("Enter item name: ").strip()
    action = {"type": action_type, "user": user, "item": item}
elif action_type == "comment":
    comment = input("Enter comment: ").strip()
    action = {"type": action_type, "user": user, "comment": comment}
else:
    action = {"type": action_type, "user": user}

# Match statement
match action:
    case {"type": "login", "user": user}:
        print(f"{user} logged in")
    case {"type": "logout", "user": user}:
        print(f"{user} logged out")
    case {"type": "purchase", "user": user, "item": item}:
        print(f"{user} purchased {item}")
    case {"type": "comment", "user": user, "comment": cmt}:
        print(f"{user} commented: {cmt}")
    case _:
        print("Unknown action")
```

Verdict: Stick to Match Statement in this case instead of Conditional Statement
