Import JSON Library

```python
import json
```

üìÅ Step 1: Create the JSON Data File

```python
{
  "location": "Dhaka",
  "date": "2026-01-05",
  "readings": [
    {"altitude": 10, "PM2_5": 95, "PM10": 140},
    {"altitude": 30, "PM2_5": 82, "PM10": 120},
    {"altitude": 50, "PM2_5": 70, "PM10": 105}
  ]
}
```

üìå Explanation

JSON stores data as key‚Äìvalue pairs

readings is a list of dictionaries

Each dictionary is one sensor measurement

üß© Step 2: Read JSON Data in Python

```python
import json

with open("air_quality.json", "r") as file:
    data = json.load(file)

print(data["location"])
```

üìå What happens here

json.load() converts JSON into a Python dictionary

You can access values using keys like data["location"]

üîç Step 3: Extract and Use the Data

```python
for reading in data["readings"]:
    print(
        "Altitude:", reading["altitude"],
        "PM2.5:", reading["PM2_5"],
        "PM10:", reading["PM10"]
    )
```

üìå Why this matters

This is how ML models and dashboards read sensor data

Works exactly like real API responses

‚úèÔ∏è Step 4: Update JSON Data

```python
new_reading = {"altitude": 70, "PM2_5": 60, "PM10": 90}
data["readings"].append(new_reading)

with open("air_quality.json", "w") as file:
    json.dump(data, file, indent=4)

```

üìå What you did

Added new sensor data

Saved updated data back to JSON

üìä Step 5: Simple Analysis (Average PM2.5)

```python
pm25_values = [r["PM2_5"] for r in data["readings"]]
average_pm25 = sum(pm25_values) / len(pm25_values)

print("Average PM2.5:", average_pm25)
```

üìå Concept used

JSON ‚Üí Python list ‚Üí simple math

This is how analytics pipelines begin
